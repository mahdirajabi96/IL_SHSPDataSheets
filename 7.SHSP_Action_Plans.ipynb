```python
def GetActionDF(df,f1,f2):
    df2 = pd.DataFrame(columns=['NObj','NStr','NAction','Obj','Str','Action'])
    Obj = 0
    Str = 0
    Act = 0
    j = 0
    for i,r in df.iterrows():
        if not pd.isnull(r[f1]):
            s = str(r[f1])
            s = s.lower()
            if s.find('objective')!=-1:
                Obj += 1 
                Str = 0
                Act = 0
                ObjTEXT = r[f2]
            if s.find('strategy')!=-1:
                Str += 1 
                StrTEXT = r[f2]
                Act = 0
            if s.find('action')!=-1:
                ActTEXT = r[f2]
                Act += 1
            #print(s,Obj,Str,Act)
        if Obj>0 and Str>0 and Act>0 and (not pd.isnull(ActTEXT) or not pd.isnull(StrTEXT) or not pd.isnull(ObjTEXT)):
            j += 1
            df2.loc[j] = [Obj,Str,Act,ObjTEXT,StrTEXT,ActTEXT]
    df2 = df2.drop_duplicates()
    return(df2)
```

```python
APDF = pd.DataFrame()
for k in ['IN','RD','TN','PD','PC','IM','SA','YD','OD','UO','TIM','IS','DF','HV','MC','WZ']:
    #print(k)
    df = pd.read_excel(AP_Dir+'\\'+k+'.xlsx',AP_Path[k][0],skiprows=AP_Path[k][1])
    df1 = GetActionDF(df,AP_Path[k][2],AP_Path[k][3])
    #print(' - Obj: {}, Str: {}, Actions: {}'.format(max(df1.NObj),len(set(df1.NStr)),df1.shape[0]))
    mi = pd.MultiIndex.from_tuples([(k,o,s,a) for o,s,a in zip(df1.NObj,df1.NStr,df1.NAction)],names = ['EA','NObj','NStr','NAction'])
    df1.index = mi
    df1 = df1[['Obj','Str','Action']]
    df1
    APDF = pd.concat([APDF,df1])
    #print(k,df1.shape)
    #display(df1)
APDF = APDF.sort_index()
APDF['NAction'] = [i[3] for i in APDF.index]
APDF = APDF.loc[APDF[(~pd.isnull(APDF.Action)) | (APDF.NAction==1)].index,['Obj','Str','Action']]
df = pd.DataFrame(columns=['TotObj','TotStr','TotActions'],index = ['IN','RD','TN','PD','PC','IM','SA','YD','OD','UO','TIM','IS','DF','HV','MC','WZ'])
for ea in df.index:
    df.loc[ea] = [len(set(APDF[~pd.isnull(APDF.Obj)].loc[ea]['Obj'])),
                  len(set(APDF[~pd.isnull(APDF.Str)].loc[ea]['Str'])),
                  len(set(APDF[~pd.isnull(APDF.Action)].loc[ea]['Action']))]
df = df.sort_values('TotActions',ascending= False)
df
```
